# WAGES OF THE PLAYERS

```python
salary = df[['short_name', 'wage_eur']]
salary.head()
```

```python
#sorting the dataset by wages
```

```python
salary = salary.sort_values(by="wage_eur", ascending=False)
salary.head()
```

```python
plt.figure(figsize=(10, 10))
plt.bar(list(salary["short_name"])[0:5], list(salary["wage_eur"])[0:5], color=["blue", "green", "red", "pink", "yellow"])
```

# Analyses on Germany

```python
Germany = df[df["nationality"]== "Germany"]
Germany.head(10)
```

# Tallest German player

```python
Tallest = Germany.sort_values(by = "height_cm", ascending= False)
Tallest.head()
```

```python
plt.figure(figsize= (10,10))
plt.bar(list(Tallest["short_name"])[0:5], list(Tallest["height_cm"])[0:5], color="g")
```
