# Extracting good charcters

```python
df1=df[df["Alignment"]=="good"]
df1.head()
```

# Charcters with high speed

```python
speed = df1[["Name", "Speed"]].sort_values(by="Speed", ascending=False).head(5)
speed
```

```python
plt.figure(figsize=(10,10))
plt.bar(list(speed["Name"][0:5]),list(speed["Speed"][0:5]), color="g")
```

# Find out all the charcters with speed 100.

```python
max_speed = df1[df1["Speed"]==100]
```

# characters with minimum speed

```python
min_speed = df1[df1["Speed"]==1]
min_speed
```

# Characters with high power

```python
power = df1.sort_values(by="Total", ascending=False)
power 
```

```python
max_power = power[power["Power"]==100]
max_power
```

```python
max_power.shape
```

There are 33 super power charcters  with maximum power = 100

```python
plt.figure(figsize=(10,10))
plt.bar(list(power["Name"][0:5]), list(power["Total"][0:5]), color="g")
```

# Analysing Bad Characters

```python
df2= df[df["Alignment"]=="bad"]
df2.head()
```

# Bad Characters with high total power

```python
high_power_bad = df2.sort_values(by="Total", ascending=False)
high_power_bad.head()
```

```python
plt.figure(figsize=(10,10))
plt.bar(list(high_power_bad["Name"][0:5]), list(high_power_bad["Total"][0:5]), color="g")
```
